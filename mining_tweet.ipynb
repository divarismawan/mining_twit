```python
num_tweets = 10000
date_since = '2019-04-07'
date_until = '2019-05-10'
query = ['traffic accident', '#kecelakaan', 'terjadi kecelakaan', '#tabrakan']
```

```python

for i in query:
    total =0
    file = open("hasil/"+i+"/{}_{}_{}.json".format(i,date_since,date_until), 'w+')
    for tweet in tweepy.Cursor(api.search,q = i, since = date_since, until = date_until).items(num_tweets):
        total +=1
        print("created_at : {} location : {}".format(tweet.created_at, tweet.user.location))
        file.write(json.dumps(tweet._json) + '\n')
    print("\n \nTotal {} : {} \n \n".format(i,total))
print("Selesai")
```
